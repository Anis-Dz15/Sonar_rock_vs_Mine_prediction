```python
# Importing packages 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Importing the Dataset 

df = pd.read_csv('Copy of sonar data.csv',header=None)
```

```python
# Overview of the dataset

df.head()
```

```python
# Preview of the dataset if it contains empty data
df.isnull().sum().any()
```

```python
# Number of Mine and Rock
df[60].value_counts()
```

```python
# Separating Data and Labels 
X = df.drop(columns=60, axis=1)
y = df[60]
```

```python
X.head()
```

```python
y.head()
```

```python
# Training and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1,stratify=y) 
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
# Application of the traning model "Logistic Regression" 
lg = LogisticRegression()
```

```python
# Training the logistic Regression model with training data

lg.fit(X_train,y_train)
```

```python
# Model evaluation

# Accuracy on training data
X_train_prediction = lg.predict(X_train)
acc_score = accuracy_score(X_train_prediction,y_train)

print('Accuracy on training data is :',acc_score *100,'%' )
```

```python
# Accuracy on test data
X_test_prediction = lg.predict(X_test)
acc_score = accuracy_score(X_test_prediction,y_test)

print('Accuracy on testing data is :',acc_score *100,'%' )
```
